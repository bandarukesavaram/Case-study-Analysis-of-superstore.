```python
import numpy as np
import pandas as pd
dataset=pd.read_csv("/content/SampleSuperstore.csv")
dataset.head()
```

```python
dataset.describe()
```

```python
dataset["State"].unique()
```

```python
state_profit=pd.DataFrame(dataset.groupby("State")["Profit"].sum())
state_profit.reset_index(inplace=True)
state_profit=state_profit.sort_values(by="Profit",ascending=False)
state_profit.head(15)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 6))
plt.bar(state_profit["State"], state_profit["Profit"],color='red')
plt.xlabel("State")
plt.ylabel("Total Profit ")
plt.title("Total profit by state")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
dataset["Category"].unique()
```

```python
dataset["Sub-Category"].unique()
```

```python
categories_info=pd.DataFrame(dataset.groupby("Category")["Profit"].sum())
categories_info.reset_index(inplace=True)
categories_profit=categories_info.sort_values(by="Profit",ascending=False)
categories_profit.head(3)
```

```python
import matplotlib.pyplot as plt
plt.bar(categories_profit["Category"],categories_profit["Profit"])
plt.xticks(categories_profit['Category'] , rotation = "vertical")
plt.xlabel("Categories")
plt.ylabel("Profit")
plt.title('Category-Wise Profit')
plt.show()
```

```python
dataset["Sub-Category"].unique()
```

```python
sub_categoriesinfo=pd.DataFrame(dataset.groupby("Sub-Category")["Profit"].sum())
sub_categoriesinfo.reset_index(inplace=True)
sub_categories_profit=sub_categoriesinfo.sort_values(by="Profit",ascending=False)
sub_categories_profit.head(10)
```

```python
import matplotlib.pyplot as plt
plt.bar(sub_categories_profit["Sub-Category"],sub_categories_profit["Profit"],color='orange')
plt.xticks(rotation =90)
plt.xlabel("Sub-Categories")
plt.ylabel("Profit")
plt.title('Sub-Category-Wise Profit')
plt.show()
```

```python
city_highest_sales = pd.DataFrame(dataset.groupby('City')['Sales'].sum())
city_highest_sales =city_highest_sales.sort_values(by="Sales",ascending=False)
city_highest_sales.head()
```
